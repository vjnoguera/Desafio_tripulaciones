```python
import openai
from pymongo import MongoClient
import pandas as pd

# Definir la función para descargar todas las colecciones desde MongoDB
def download_all_collections_from_mongodb(uri, db_name):
    # Conectar a MongoDB
    client = MongoClient(uri)
    try:
        # Seleccionar la base de datos
        db = client[db_name]
        
        # Listar todas las colecciones
        collection_names = db.list_collection_names()

        # Descargar los datos de todas las colecciones
        dataframes = {}
        for collection_name in collection_names:
            collection = db[collection_name]
            data = list(collection.find())
            df = pd.DataFrame(data)
            dataframes[collection_name] = df
    finally:
        # Cerrar la conexión
        client.close()
    
    return dataframes

# Datos de conexión a MongoDB
uri = "mongodb+srv://Cluster78620:uuMda3E0UnvIPzdi@cluster786∫20.jcfu9lt.mongodb.net"
db_name = "Samoo"

# Descargar todas las colecciones y convertirlas en DataFrames
dataframes = download_all_collections_from_mongodb(uri, db_name)

# Combinar todos los DataFrames en uno solo
combined_df = pd.concat(dataframes.values(), ignore_index=True)

# Configurar la API de OpenAI
openai.api_key = "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" # Reemplaza 'tu_api_key_de_openai' con tu clave API real

# Crear el prompt para ChatGPT
question = "¿De qué ciudades vienen los asistentes?"
prompt = f"""En base al siguiente DataFrame {combined_df}, quiero que respondas a la pregunta registrada en "{question}". En la respuesta, intenta ser lo más conciso posible y pon solo texto normal, nada parecido a "\\n" o similar."""

# Llamar a la API de OpenAI
response = openai.Completion.create(
    engine="text-davinci-003",  # Puedes ajustar el modelo según tus necesidades
    prompt=prompt,
    max_tokens=250,
    temperature=0.01
)

# Obtener la respuesta generada
nav = response.choices[0].text.strip()
print(nav)
```
