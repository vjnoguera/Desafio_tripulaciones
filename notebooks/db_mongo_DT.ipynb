```python
import pandas as pd
import random

# Listas proporcionadas
alergias = ["Gluten", "Lácteos", "Huevos", "Frutos secos", "Moluscos"]
food_preferences = ["Vegetariano", "Vegano", "Carnívoro"]
intereses = ["Tecnología", "Gestión de Proyectos", "Agile", "Softskills", "Marketing Digital", "Negocios", "Emprendimiento", "Educación", "Formación", "Salud y Bienestar", "Creatividad", "Diseño"]
Posición = [
    "Desarrollador de Contenidos", "Instructor / Formador", "Coordinador de Cursos", 
    "Administrador de Plataforma", "Especialista en Evaluación", "Diseñador Instruccional", 
    "Director de Formación", "Consultor de E-learning", "Gestor de Proyectos", 
    "Desarrollador de E-learning", "Investigador en E-learning", "Especialista en Soporte Técnico", 
    "Gestor de Comunidad", "Responsable de Calidad", "Director de Tecnología", 
    "Analista de Datos Educativos", "Especialista en Marketing de E-learning", 
    "Facilitador de Talleres", "Asesor Pedagógico", "Director de Innovación"
]

# Función para generar la columna de alergias
def generate_allergies():
    if random.random() < 0.01:  # 1% de probabilidad de tener 2 alergias
        return random.sample(alergias, 2)
    elif random.random() < 0.5:  # 50% de probabilidad de tener 1 alergia
        return random.sample(alergias, 1)
    else:
        return []

# Función para generar la columna de food_preferences
def generate_food_preference():
    return random.choice(food_preferences)

# Función para generar la columna de intereses
def generate_interests():
    return random.sample(intereses, random.randint(3, 5))

# Crear el DataFrame con las 4 columnas y 300 filas
df = pd.DataFrame({
    'Posición': [random.choice(Posición) for _ in range(300)],
    'Alergias': [generate_allergies() for _ in range(300)],
    'Food Preferences': [generate_food_preference() for _ in range(300)],
    'Intereses': [generate_interests() for _ in range(300)]
})

df
```
